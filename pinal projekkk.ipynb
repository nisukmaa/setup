```python
X=train_coba.drop(columns='SalePrice')
y=np.log(train_coba.SalePrice)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
im=LinearRegression()
```

```python
model=im.fit(X_train,y_train)
```

```python
im.intercept_
```

```python
im.coef_
```

```python
coeff_df=pd.DataFrame(im.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
print ("R^2 = ", model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE = ', mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=1,
            color='r')
plt.xlabel('Prediksi Harga')
plt.ylabel('Harga sebenernya')
plt.title('Linear Regression Model')
plt.show()
```
